```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('heart_disease_dataset.csv')
print("Primeras filas del DataFrame:")
print(df.head())
```

```python
resultados = []

filtered_df = df[(df['Heart Disease'] == 1) & (df['Cholesterol'] > 200) & (df['Blood Pressure'] > 130)]
```

```python
for index, row in filtered_df.iterrows():
    resultados.append({
        'Age': row['Age'],
        'Gender': row['Gender'],
        'Cholesterol': row['Cholesterol'],
        'Blood Pressure': row['Blood Pressure'],
        'Stress Level': row['Stress Level'],
        'Heart Disease': row['Heart Disease'],
        'Alcohol Intake': row['Alcohol Intake'],
        'Smoking': row['Smoking']
    })

# Mostrar el contenido de la lista resultados
print("Contenido de la lista resultados:")
for record in resultados:
    print(record)
```

```python
# Conectar a la base de datos SQLite (o crearla si no existe)
conn = sqlite3.connect('heart_disease.db')
cursor = conn.cursor()
```

```python
try:
    create_table = """
    CREATE TABLE IF NOT EXISTS heart_disease_data (
        Age INT,
        Gender NVARCHAR(50),
        Cholesterol FLOAT,
        Blood_Pressure FLOAT,
        Stress_Level INT, 
        Heart_Disease INT,
        Alcohol_Intake NVARCHAR(50), 
        Smoking NVARCHAR(50)  
    );
    """
    
    cursor.execute(create_table)
    conn.commit()
    print("Tabla creada (si no existía).")

except Exception as e:
    print(f"Error occurred: {e}")
```

```python
# Insertar los datos filtrados en la base de datos
try:
    for record in resultados:
        cursor.execute("""
        INSERT INTO heart_disease_data (
            Age, 
            Gender, 
            Cholesterol, 
            Blood_Pressure,
            Stress_Level,
            Heart_Disease,
            Alcohol_Intake,
            Smoking,  
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """,
         record['Age'], record['Gender'], record['Cholesterol'],
          record['Blood_Pressure'], record['Stress_Level'], record['Heart_Disease'],
           record['Alcohol_Intake'], record['Smoking'],
        )
        conn.commit()
        print("Datos insertados en la tabla.")
except Exception as e:
    print(f"Error occurred during data insertion: {e}")
finally:
    # Cerrar la conexión
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```

```python
# Verificar el contenido de la tabla
try:
    conn = sqlite3.connect('heart_disease.db')
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM heart_disease_data")
    rows = cursor.fetchall()
    print("Contenido de la tabla heart_disease_data:")
    for row in rows:
        print(row)
    
except Exception as e:
    print(f"Error occurred during data retrieval: {e}")
```
